```python
df = df.drop_duplicates()    #drop duplicates
df    #saves new dataframe
```

```python
df = df.drop(columns = "Not_Useful_Column")    #drops a column, 'saves df' or 'inplace=True' is specified, it means that the operation will modify the original DataFrame itself
df
```

```python
#df["Last_Name"] = df["Last_Name"].str.strip("...")    #before = saves to the df["Last_Name"]
#df["Last_Name"] = df["Last_Name"].str.strip("/")      #after = strips character from column
#df["Last_Name"] = df["Last_Name"].str.strip("_")
df["Last_Name"] = df["Last_Name"].str.strip("123._/")   #You can put all characters together
df
```

```python
#get rid of all non numeric data in column
#Change entire column to strings
#df ["Phone_Number"] = df ["Phone_Number"].str.replace("[^a-zA-Z0-9]", "")
#inside the bracket says replace all characters 'except these characters a-zA-Z0-9' and replace them with nothing
###df ["Phone_Number"].apply(lambda x: x[0:3] + "-" + x[3:6] + "-" + x[6:10])  #formats the 1st 3 strings then the next and the next. 
###Above gets error needs to be a string
#df ["Phone_Number"] = df ["Phone_Number"].apply(lambda x: str(x)) #applys lambda to each value of x making it a string
#df ["Phone_Number"] = df ["Phone_Number"].apply(lambda x: x[0:3] + "-" + x[3:6] + "-" + x[6:10])
#Above formats the string(number)
#df ["Phone_Number"] = df ["Phone_Number"].str.replace('nan--', '')
df ["Phone_Number"] = df ["Phone_Number"].str.replace('Na--', '')
df
```

```python
df[["Street Address","State", "Zip_Code"]]=df["Address"].str.split(',',2,expand=True)
#Above is creating different columns. They are appended and do not replace Address
df
```

```python
df["Paying Customer"] = df["Paying Customer"].str.replace('Yes', 'Y')
df["Paying Customer"] = df["Paying Customer"].str.replace('No', 'N')
df["Do_Not_Contact"] = df["Do_Not_Contact"].str.replace('Yes', 'Y')
df["Do_Not_Contact"] = df["Do_Not_Contact"].str.replace('No', 'N')
df
```

```python
#df = df.replace('N/a','')
df = df.fillna('') #This fills all Nan(not a number) with a blank
df
```

```python
#remove all rows with Y in Do_Not_Contact column
for x in df.index:
    if df.loc[x, "Do_Not_Contact"] == 'Y':
          df.drop(x, inplace=True)
df
```

```python
#If the column has a blank get rid of it
for x in df.index:
    if df.loc[x, "Phone_Number"] == '':
          df.drop(x, inplace=True)
df
#Another way to drop null values
#df.dropna(subset="Phone_Number"), inplace=True
```
